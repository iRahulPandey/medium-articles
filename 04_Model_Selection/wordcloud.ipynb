```python
from wordcloud import WordCloud
from os import path, getcwd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
supervised_models = """Linear Regression
Logistic Regression
Decision Tree
SVM
Naive Bayes
kNN
K-Means
Random Forest
Dimensionality Reduction Algorithms
Gradient Boosting algorithms
GBM
XGBoost
LightGBM
CatBoost"""

unsupervised_models = """K-means for clustering problems
Apriori algorithm for association rule learning problems
Principal Component Analysis
Singular Value Decomposition
Independent Component Analysis"""

reinforcementleaning_models = """Monte Carlo
Q-learning
SARSA
Q-learning - Lambda
SARSA - Lambda
DQN
DDPG
A3C
"""

house = """per capita crime rate by town
proportion of residential land zoned for lots over 25,000 sq.ft.
proportion of non-retail business acres per town.
Charles River dummy variable (1 if tract bounds river; 0 otherwise)
nitric oxides concentration (parts per 10 million)
average number of rooms per dwelling
proportion of owner-occupied units built prior to 1940
weighted distances to five Boston employment centres
index of accessibility to radial highways
full-value property-tax rate per $10,000
pupil-teacher ratio by town
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
lower status of the population
Median value of owner-occupied homes in $1000's"""
```

```python
# makes the circle using numpy
x, y = np.ogrid[:300, :300]
mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)

# Create and generate a word cloud image:
wordcloud = WordCloud(background_color="white",
 mask=mask, contour_width=0.0, contour_color="black")
wordcloud.generate(supervised_models)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud(background_color="white", mask=mask ,contour_width=0.0, contour_color="black")
wordcloud.generate(unsupervised_models)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
